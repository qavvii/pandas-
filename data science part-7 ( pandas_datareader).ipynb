```python
import pandas as pd 
from pandas_datareader import data as web
import datetime as dt 
amzn=web.DataReader(name='AMZN',data_source='stooq',start='2017-08-25',end=dt.datetime.today())
amzn
```

```python
amzn2=web.DataReader('AMZN',data_source='stooq',start='2017-08-10',end='2017-08-24')
amzn2
```

```python
amzn_complete=amzn.append(amzn2)
```

```python
amzn_complete
```

```python
amzn_complete.sort_index(inplace=True,ascending=True)
amzn_complete
```

```python
amzn_complete2=pd.concat([amzn,amzn2])
amzn_complete2
```

```python
amzn_complete2.sort_index(ascending=True,inplace=True)
amzn_complete2
```

```python
MSFT=web.DataReader('AMZN',data_source='stooq',start='2017-08-10',end='2017-08-24')
MSFT
```

```python
MSFT=web.DataReader('MSFT',data_source='stooq',start='2017-08-24',end=dt.datetime.now())
MSFT1=MSFT[['Open','High']]
MSFT1
```

```python
MSFT2=MSFT[['Low','Close','Volume']]
MSFT2
```

```python
MSFT_merge=MSFT1.merge(MSFT2,left_index=True,right_index=True)
MSFT_merge
```

```python
#another way 
msft_merge3=pd.merge(MSFT1,MSFT2,how='inner',left_index=True,right_index=True)
msft_merge3
#optional how='inner '
```
